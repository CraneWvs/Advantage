Exploration of geoJson data

```python
import json
```

## Taxi zones

```python
with open('../data-ingestions/nyc-taxi-zones.geojson', 'r') as fd:
    data = json.load(fd)
print(data.keys())
```

```python
data.get('type')
```

```python
features = data.get('features')
```

```python
print(features[0])
```

```python
print(len(features))
```

```python
for feature in features:
    print(len(feature['properties']['borough']))
```
